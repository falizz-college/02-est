```python
import matplotlib.pyplot as plt

nfilhos=[ dados['N de Filhos']]
plt.figure(figsize=(4, 3))
plt.xlabel('Quantidade de filhos')
plt.ylabel('Frequência Absoluta')
plt.hist(nfilhos, 5, rwidth=0.9)
plt.grid(True)
plt.show()

salario=[ dados['Salario (x Sal Min)']]
plt.figure(figsize=(4, 3))
plt.xlabel('Salário (x Salário Mínimo)')
plt.ylabel('Frequência Absoluta')
plt.hist(salario, 25, rwidth=0.9)
plt.grid(True)
plt.show()

idades=[ dados['Idade (em anos)']]
plt.figure(figsize=(4, 3))
plt.xlabel('Idade')
plt.ylabel('Frequência Absoluta')
plt.hist(idades, 25, rwidth=0.9)
plt.grid(True)
plt.show()
```

Sim, é possível calcular a frequência com que as variáveis qualitativas aparecem

```python
estadocivil=[ dados['Estado Civil']]
plt.figure(figsize=(5, 3))
plt.ylabel('Frequência Absoluta')
plt.hist(estadocivil, rwidth=1, bins=3)
plt.grid(True)
plt.show()

escolaridade=[ dados['Grau de Instrução']]
plt.figure(figsize=(5, 3))
plt.ylabel('Frequência Absoluta')
plt.hist(escolaridade, rwidth=1, bins=5)
plt.grid(True)
plt.show()

regiao=[ dados['Região de Procedência']]
plt.figure(figsize=(5, 3))
plt.ylabel('Frequência Absoluta')
plt.hist(regiao, rwidth=1, bins=5)
plt.grid(True)
plt.show()
```

```python
dados.boxplot(column= 'N de Filhos')
plt.show()
dados.boxplot(column= 'Salario (x Sal Min)')
plt.show()
dados.boxplot(column= 'Idade (em anos)')
plt.show()
```

Sim, através do histograma podemos ver quais dados aparecem com mais frequência e isso auxilia na percepção de outliers e através do boxplot conseguimos ver onde a mediana se encontra e se existem mais dados com valor acima da mediana, ou mais dados com valor abaixo, ou seja, quanto mais centralizada a mediana está no boxplot menos outliers existem.
